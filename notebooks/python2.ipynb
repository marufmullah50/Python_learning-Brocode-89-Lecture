```python
#birthday song print
#Without Parameters
def congratulate():
    print("Congratulations on your special day!")
    print("Wishing you all the best!")
    print("\n")
#With Parameters

def happy_birthday(name):
    print("Happy Birthday to You")
    print(f"Happy Birthday Dear {name}")
    print("You are Old")
    print()

# Example usage
congratulate()
happy_birthday("Alice")

#Multuple Parameters
def happy_birthday(name ,age):
    print("Happy Birthday to You")
    print(f"Happy Birthday Dear {name}")
    print(f"You are {age} years old")
    print()
print("With Multiple Parameters:\n")

happy_birthday("Bob", 30) # Parameters order matters

```

```python
#Return Statement 
def add(x,y):
    z=x+y
    return z
def multiply(a,b):
    c=a*b
    return c
def subtract(m,n):
    p=m-n
    return p
def divide(q,r):

    s=q/r
    return s

print("Addition:",add(5,3)) # here 5 and 3 are arguments become 8
print("Multiplication:",multiply(4,2)) # here 4 and 2 are arguments become 8
print("Subtraction:",subtract(10,6)) # here 10 and 6 are arguments become 4
print("Division:",divide(20,4)) # here 20 and 4 are arguments become 5.0
```

```python
def calculate_area(length, width):
    return length*width
print(f"Rectangular Area = {calculate_area(3,4)}")

def triangular_area(height, base):
    return 0.5*height*base
tri = triangular_area(3,4)
print(f"Triangular Area = {tri}")

def netprice(list_price , discount = 0 , tax = 0.05): #deafult value for tax and discount

    return list_price*(1-discount)*(1+tax)

print(f"using both deafult valur {netprice(500)}") # here discount = 0 ,tax = 0.05  as deafult value
print(f"using only one deafult value : {netprice(500,0.1)}") # here 0.1 is discount
print(f"no deafult value giving each one a value : {netprice(500 , 0.1 , 0)} ") # no deafult value
```

```python
def adding(*args):
    total =0
    for arg in args :
        total +=arg
    return total
x= adding(1,2,3,4,5) # can do multiple 
print(x)
```

#**kwargs

```python
#add as many as i want

def adds(**kwargs):
    print(type(kwargs))
adds()
```

```python
#Getting values
def print_address(**kwargs):
    for value in kwargs.values():
        print(value)


print_address(street="Railgate1" ,
              city= "Narsingdi",
              State= "Dhaka",
              Country ="Bangladesh")
```

```python
#Getting keys

def print_address(**kwargs):
    for key in kwargs.keys():
        print(key)


print_address(street="Railgate1" ,
              city= "Narsingdi",
              State= "Dhaka",
              Country ="Bangladesh")
```

```python
# Getting both

def print_address(**kwargs):
    for key ,value in kwargs.items():
        print(f"{key}: {value}")


print_address(street="Railgate1" ,
              city= "Narsingdi",
              State= "Dhaka",
              Country ="Bangladesh")
```

#*args and **kwargs together


```python
def shipping_label(*args ,**kwargs):
    print("Name:")
    for arg in args:
        print(arg, end=" ")
    print("\n")
    print("Address:")
    for key,value in kwargs.items():
        print(f"{key}: {value}" )
shipping_label("Md.","Maruf","Mullah",
              street="Railgate1" ,
              city= "Narsingdi",
              State= "Dhaka",
              Country ="Bangladesh")
               
               
```

```python
#Without second for loop

def shipping_label(*args ,**kwargs):
    print("Name:")
    for arg in args:
        print(arg, end=" ")
    print("\n")
    print("Address:")
    print(f"{kwargs.get('street', '')}, {kwargs.get('city', '')}")
    print(f"{kwargs.get('State','')},{}")
shipping_label("Md.","Maruf","Mullah",
              street="Railgate1" ,
              city= "Narsingdi",
              State= "Dhaka",
              Country ="Bangladesh")
               
               
```

```python
numbers = [1,-2,4,2,5,-6,-3,10,-8]
positive_nums=[num for num in numbers if num >=0]
negative_nums = [num for num in numbers if num <0]
even_nums = [num for num in numbers if num%2==0]
odd_nums = [num for num in numbers if num%2!=0]

print("Postive Numbers:",positive_nums)
print("Negative Numbers:",negative_nums)
print("Even Numbers:",even_nums)
print("Odd Numbers",odd_nums)
```

#Match case

```python
def week(day):
    if day==1:
        return "Sunday"
    elif day == 2:
        return "Monday"
    elif day ==3:
        return "Tuesday"
    elif day ==4:
        return "Wednesday"
    elif day == 5:
        return "Thursday"
    elif day ==6:
        return "Friday"
    elif day ==7:
        return "Saturday"
    else:
        return "Not a valid day"
week(6)
    
```
