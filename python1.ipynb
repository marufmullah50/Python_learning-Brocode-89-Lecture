```python
#Indexing
department ="Mechanical Engineering"
print(department[0])  #M
print(department[5])  #n
print(department[-1]) #g
print(department[3:6]) # [start:end]
#last 3 characters
print(department[-3:])#Lat 3
print(department[-3:-1:1]) # [start:end:step] last digit skipped
```

#Loop


```python
#While Loop
i = int(input("Enter a number of  times name to be printed:"))
while i>0:
    print("Maruf")
    i -=1
    
```

```python
#For loop
i = int(input("Enter a number of  times name to be printed:"))
j=0
for j in range(i):
    print("Maruf")
    i -=1
```

```python
for x in range(0,6):
    print("Maruf")
    print(x)
```

```python
for x in range(0,60,10): # for x in range(start, end, step)
    print("Maruf")
    print(x)
```

```python
i = int(input("Enter number of times the name will be printed: "))

# Simulating do-while loop
#Break condition at the end
while True:
    print("Maruf")
    i -= 1
    if i < 1:
        break
```

```python
#Continue Statement
for i in range(1, 8):
    if i == 5:
        continue
    print("Maruf")
```
