#Multiple Inheritance = inherit from more than one parent class ,
                          
                           C(A,B)

#Multilevel inheritance = inherit from a parent which inherits from another parent 

                        C(B) <- B(A) <- A
                    

```python
class Predator:
    def hunt(self):
        print("This animal Hunt")
class Prey:
    def flee(self):
        print("This Animal Flee")

class Rabbit(Prey):
    pass
class Hawk(Predator):
    pass
class Fish(Predator,Prey):
    pass

rabbit = Rabbit()
hawk=Hawk()
fish =Fish()

print("Rabbit")
rabbit.flee()
print("\n Hawk")
hawk.hunt()
print("\nFish")
fish.hunt()
fish.flee()
```

#Multilevel

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def eat(self):
        print(f"This {self.name} is eating")
    def sleep(self):
        print(f"This {self.name} is sleeping")

class Predator(Animal):
    def hunt(self):
        print(f"This {self.name} Hunt")
class Prey(Animal):
    def flee(self):
        print(f"This {self.name} Flee")

class Rabbit(Prey): # it inherit from parent which inherit from another parent
    pass             # Grand parent level inheritence
class Hawk(Predator):
    pass
class Fish(Predator,Prey):
    pass

rabbit = Rabbit("Bunny")
hawk=Hawk("Tony")
fish =Fish("Nemo")
rabbit.eat()
fish.hunt()

```

#Abstract Class: A class that cannot be instantiated on its own; Meant to be subclassed. They cannot abstract methods , which are declared but have no implementation.

#Abstract classes Benefits:

1.Prevents instantiation of the class itself

2.Requires children to use inherited abstract methods

```python
from abc import ABC,abstractmethod

class vehicle(ABC):
    @abstractmethod
    def go(self): #Method to inherit for child thats why @abstractmethod used to remember
        print("You shall pass")
    @abstractmethod
    def stop(self):
        print("You must stop")
class Boat(vehicle):
    def go(self): 
        print("You shall pass")
    def stop(self):
        print("You must stop")

class Bike(vehicle):
    def stop(self):
        print("You must stop")
        #Here now if i try bike=Bike() it will say error
        #Cause i didn't include go stop both like parent class

boat =Boat()
boat.go()
boat.stop()
```

#super() = Function in a child class to call methods from a parent class (superclass).

#Allows you to extend the functionality of the inherited methods

```python
class Shape:#Parent
    def __init__(self,color,filled):
        self.color =color
        self.filled = filled #Using super in child class the i won't require to initialize it

class Circle(Shape):
    def __init__(self,color,filled,radius):
        super().__init__(color,filled)
        self.radius =radius

class Square(Shape):
    def __init__(self,color,filled,width):
        super().__init__(color,filled)
        self.width =width

class Triangle(Shape):
    def __init__(self,color,filled,width,height):
        super().__init__(color,filled)
        self.width =width
        self.height =height

circle =Circle("red","filled",5)
print(f"circle color is:{circle.color} , it is {circle.filled} and the radius is {circle.radius}cm")

sq=Square("red","not filled",10)
print(f"Square color is:{sq.color} , it is {sq.filled} and the width is {sq.width}cm")


tri=Triangle("red","notfilled",10,20)
print(f"Triangle color is:{tri.color} , it is {tri.filled}, the width is {tri.width}cm and height is {tri.height}cm")

```
